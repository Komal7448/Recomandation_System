# **Task 5: Hyperparameter Tuning**

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
```

```python
#Defining hyperparameter tuning space

param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']}
```

```python
#Performing hyperparameter tuning using GridSearchCV

clf = DecisionTreeClassifier(random_state=42)
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='f1_macro')
grid_search.fit(X_train, Y_train)
```

```python
GridSearchCV(cv=5, estimator=DecisionTreeClassifier(random_state=42),
             param_grid={'criterion': ['gini', 'entropy'],
                         'max_depth': [None, 5, 10, 15],
                         'min_samples_split': [2, 5, 10]},
             scoring='f1_macro')
```

```python
#Getting the best hyperparameters and the corresponding model

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```

```python
best_params
```

```python

best_model
```

```python
#Evaluating the best model on the testing set

Y_pred = best_model.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='macro')
recall = recall_score(Y_test, Y_pred, average='macro')
f1 = f1_score(Y_test, Y_pred, average='macro')

# Calculate ROC-AUC score for multiclass problems
y_pred_proba = best_model.predict_proba(X_test)
roc_auc = roc_auc_score(Y_test, y_pred_proba, multi_class='ovr')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("ROC-AUC:", roc_auc)
```


# **1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance?**
Decision Tree models come with several hyperparameters that significantly impact their accuracy, complexity, and generalization ability. The most common ones are:

max_depth: This controls how deep the tree can grow. Limiting the depth prevents overfitting. A tree that is too deep may memorize the training data and perform poorly on unseen data, whereas a shallow tree might underfit.

min_samples_split: The minimum number of samples required to split an internal node. Higher values reduce model complexity and improve generalization by preventing the model from learning from small, noisy splits.

min_samples_leaf: The minimum number of samples that a leaf node must have. This helps avoid creating leaf nodes that capture outliers or noise.

criterion: This defines the function used to measure the quality of a splitâ€”commonly gini for Gini impurity or entropy for information gain. Both work similarly, though entropy is more computationally expensive.

max_features: The number of features to consider when searching for the best split. Restricting this adds randomness to the model and is useful in ensemble methods like Random Forest.

In this project, hyperparameters like max_depth, min_samples_split, and criterion were tuned using GridSearchCV. The model with the best combination achieved a balanced performance by avoiding both underfitting and overfitting.

# **2. What is the difference between Label Encoding and One-Hot Encoding?**
Label Encoding and One-Hot Encoding are two common techniques to convert categorical data into numerical form:

Label Encoding assigns each unique category in a feature to a different integer. For example, â€œmaleâ€ might become 0 and â€œfemaleâ€ becomes 1. It is most useful for ordinal data, where the values have a meaningful order.

One-Hot Encoding, on the other hand, creates a new binary column for each category. Each observation gets a 1 in the column for its category and 0 in all others. This is ideal for nominal data, where categories have no order (e.g., types of chest pain or thalassemia).

In this project:

Label Encoding was used for features like sex, cp, fbs, restecg, exang, slope, and thal since Decision Trees can naturally split on categorical values even when encoded numerically.

One-Hot Encoding was also explored during feature engineering for use in pipelines or with other algorithms like Logistic Regression.

In summary:

Use Label Encoding when the algorithm can handle integer-labeled categories (like Decision Trees).

Use One-Hot Encoding when the algorithm is sensitive to numerical order (like Logistic Regression or SVM) and the categories have no order.
